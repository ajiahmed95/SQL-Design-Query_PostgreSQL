```python
# Fungsi untuk membuat data users

def generate_user_data(n_data, old_table, city_table, is_print=True):
    """
    Generates new user data based on an existing table and a city table.

    Args:
        n_data: Number of new users to generate.
        old_table: Existing user table as a DataFrame.
        city_table: City table as a DataFrame.
        is_print: Whether to print the generated DataFrame.

    Returns:
        A DataFrame containing the generated user data.
    """

    new_data = {
        'user_id': old_table['user_id'],
        'username': old_table['username'],
        'first_name': old_table['first_name'],
        'last_name': old_table['last_name'],
        'email': old_table['email'],
        'phone_number': old_table['phone_number'],
        'kota_id': [random.choice(city_table['kota_id']) for i in range(n_data)]
    }

    df_user_new = pd.DataFrame(new_data)

    return df_user_new
```

```python
df_user_new = generate_user_data(n_data = 50, old_table = df_user, city_table = df_city, is_print=True)

df_user_new
```

```python
def generate_ads_data(n_data, user_table, product_table, color, transmission, is_print=True):
    """
    Generates new user data based on an existing table and a city table.

    Args:
        n_data: Number of new users to generate.
        user_table: User table as a DataFrame
        city_table: City table as a DataFrame.
        color: Color table as a DataFrame.
        transmission: Transmission table as a DataFrame
        is_print: Whether to print the generated DataFrame.

    Returns:
        A DataFrame containing the generated ads data.
    """
    # definisikan awal tanggal
    start_date = datetime(2022, 1, 1)

    # definisikan akhir tanggal
    end_date = datetime(2023, 12, 31, 23, 59, 59)

    new_data = {
        'ads_id': [i+1 for i in range(n_data)],
        'user_id': [random.choice(user_table['user_id']) for i in range(n_data)],
        'product_id': [random.choice(product_table['product_id']) for i in range(n_data)],
        'transmission_id': [random.choice(transmission['transmission_id']) for i in range(n_data)],
        'color_id': [random.choice(color['color_id']) for i in range(n_data)],
        'range_used_km': [FAKER.random_int(10000, 100000, 100) for i in range(n_data)],
        'bid_available': [FAKER.boolean(chance_of_getting_true = 80) for i in range(n_data)],
        'created_at': [FAKER.date_time_between(start_date = start_date, end_date = end_date) for i in range(n_data)]
    }

    df_ads = pd.DataFrame(new_data)

    return df_ads
```

```python
def generate_buyer_data(n_data, old_table, is_print=True):
    """
    Generates new user data based on an existing table and a city table.

    Args:
        n_data: Number of new users to generate.
        user_table: User table as a DataFrame
        city_table: City table as a DataFrame.
        color: Color table as a DataFrame.
        transmission: Transmission table as a DataFrame
        is_print: Whether to print the generated DataFrame.

    Returns:
        A DataFrame containing the generated ads data.
    """

    lat_log = [FAKER.local_latlng(country_code = 'ID', coords_only = True) for i in range(n_data)]

    new_data = {
        'buyer_id': [i+1 for i in range(n_data)],
        'first_name': old_table['first_name'],
        'last_name': old_table['last_name'],
        'email': old_table['email'],
        'phone_number': old_table['phone_number'],
        'latitude': [(f'{lat_log[i][0]}') for i in range(n_data)],
        'longitude': [(f'{lat_log[i][1]}') for i in range(n_data)],
        'username': old_table['username'],


    }

    df_buyer = pd.DataFrame(new_data)

    return df_buyer
```

```python
def generate_bid_data(n_data, buyer_df, merged_df, is_print=True):
    """
    Generates new user data based on an existing table and a city table.

    Args:
        n_data: Number of new users to generate.
        buyer_df: Buyer dataframe for generate buyer_id
        merged_df: Merged dataframe of ads and product.
        is_print: Whether to print the generated DataFrame.

    Returns:
        A DataFrame containing the generated bid_detail data.
    """

    # definisikan awal tanggal
    start_date = datetime(2022, 1, 1)

    # definisikan akhir tanggal
    end_date = datetime(2023, 12, 31, 23, 59, 59)

    new_data = {
        'bid_id': [i+1 for i in range(n_data)],
        'buyer_id': [random.choice(buyer_df['buyer_id']) for i in range(n_data)],
        'date_bid': sorted([FAKER.date_time_between(start_date=start_date, end_date=end_date) for i in range(n_data)]),
        'bid_price': [],
        'bid_status': [random.choice(['Sent', 'Accepted', 'Rejected']) for i in range(n_data)],
        'ads_id': []
    }

    for i in range(n_data):
        ad = random.choice(merged_df[['ads_id', 'price']].values)
        ads_id, actual_price = ad
        bid_price = actual_price - FAKER.random_int(1000000, 20000000, 1000000)  # Ensure bid_price <= actual price
        new_data['ads_id'].append(ads_id)
        new_data['bid_price'].append(bid_price)

    df_bid = pd.DataFrame(new_data)

    return df_bid
```
